```python
import pandas  as pd
data = pd.read_csv("../intermediate_python_datasets/gapminder.csv", index_col=0)
print(data)
```

```python
# Import package
import matplotlib.pyplot as plt
import numpy as np

pop = list(data['population']/1000000)
life_exp = list(data.life_exp)


# Build Scatter plot
plt.scatter(pop, life_exp)

# Show plot
plt.show()
```
